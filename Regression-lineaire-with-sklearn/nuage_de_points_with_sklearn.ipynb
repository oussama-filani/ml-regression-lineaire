```python
# Nuage de points with sklearn

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Définir une graine aléatoire pour la reproductibilité
np.random.seed(42)

# Génération des données
n_points = 100
X = np.random.rand(n_points) * 10  # Valeurs X entre 0 et 10
noise = np.random.randn(n_points) * 2  # Bruit gaussien
y = 2 * X + 3 + noise  # Droite y = 2x + 3 + bruit

# Affichage du nuage de points
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Données')

# Préparation des données pour sklearn
X_reshape = X.reshape(-1, 1)

# Création et entraînement du modèle
model = LinearRegression()
model.fit(X_reshape, y)

# Prédiction
y_pred = model.predict(X_reshape)

# Affichage de la droite de régression
plt.plot(X, y_pred, color='red', linewidth=2, label='Régression linéaire')

# Paramètres du graphique
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Régression linéaire sur nuage de points')
plt.grid(True)
plt.legend()

# Affichage des coefficients du modèle
print("Coefficient (pente):", model.coef_[0])
print("Ordonnée à l'origine:", model.intercept_)

# Calcul des métriques d'erreur
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Affichage des métriques
print("Erreur absolue moyenne (MAE):", mae)
print("Erreur quadratique moyenne (MSE):", mse)
print("Racine de l'erreur quadratique moyenne (RMSE):", rmse)
print("Coefficient de détermination (R²):", r2)

plt.show()
```
